```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
                    '姓名': ['张飞','关羽','刘备','典韦','许褚'],
                    '语文': [68,95,98,90,80],
                    '数学': [65,76,86,88,90],
                    '英语': [30,98,88,77,90]})
```

```python
pd.set_option('display.unicode.ambiguous_as_wide', True)
pd.set_option('display.unicode.east_asian_width', True)
pd.set_option('display.width', 180) # 设置打印宽度

def res(subjects):
    res_subjects = {'Results':['Avr','Min','Max','Var','Std']}
    for subject in subjects:
        avrS = df[subject].mean()
        minS = df[subject].min()
        maxS = df[subject].max()
        varS = df[subject].var()
        stdS = df[subject].std()
        res_subject = np.array([avrS,minS,maxS,varS,stdS])
        res_subjects[subject] = res_subject
    print('The resulf of the all subjects: \n')
    print(pd.DataFrame(res_subjects).round(1))
        
res(['语文','数学','英语'])
```

```python
def rank():
    rank_all = {'姓名': ['张飞','关羽','刘备','典韦','许褚']}
    res_all = []
    i=0
    while i<len(rank_all['姓名']):
        note = df.iloc[i,1:].sum()
        res_all.append(note)
        i+=1
    rank_all['Results'] = res_all
    print('The resulf of the all students: \n')
    pdRes = pd.DataFrame(rank_all).sort_values('Results',ascending=False)
    print(pdRes.reset_index(drop=True))
        
rank()
```

```python
df_car = pd.read_csv('car_complain.csv')
df_car.head()
```

```python
df2 = df_car.copy()
df3 = df2['problem'].str.split(',',expand=True)
df4 = pd.concat([df2,df3], axis=1)
df4.dropna(axis=1, how='all').replace('', None, regex=True)
```

```python
counts = []
for row in range(0,600):
    count = df4.iloc[row,8:16].value_counts(dropna=True).sum()-1
    counts.append(count)
#counts
dfbasic =df4[['id','brand']]
dfcounts = pd.DataFrame({'Problem Counts':counts})
dfcounts
pd.concat([dfbasic,dfcounts], axis=1)
```
